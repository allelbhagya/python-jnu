```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
y = [10,20,40,15]
fig, ax = plt.subplots()
ax.plot(x, y, marker='o')
for i in range(len(x)):
    ax.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5,-5))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Sales': [143, 120, 255, 310, 150]
}

df = pd.DataFrame(data)

# Line Plot
plt.plot(df['Year'], df['Sales'])
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Data
labels = ['mid-sems', 'end sem', 'research paper', 'programming assignment']
sizes = [25, 35, 20, 20]

# Pie Chart
plt.pie(sizes, labels=labels)
plt.title('Mark Distribution')
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

x1 = [10,20,30,40,50]
y1 = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='r')
plt.scatter(x1, y1, color='g')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python

data_bar = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [100, 200, 150, 300]
}

df_bar = pd.DataFrame(data_bar)

plt.bar(df_bar['Product'], df_bar['Sales'], color='green')
plt.title('Product Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

```

```python

data_custom = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Revenue': [300, 400, 500, 600, 700],
    'Expenses': [200, 250, 300, 400, 450],
    'Income':[100,200,300,400,500],
}

df_custom = pd.DataFrame(data_custom)

plt.plot(df_custom['Month'], df_custom['Revenue'], label='plot 1', color='blue', linestyle='-', marker='o')

plt.plot(df_custom['Month'], df_custom['Expenses'], label='plot 2', color='red', linestyle='--', marker='x')

plt.plot(df_custom['Month'], df_custom['Income'], label='plot 3', color='black', linestyle='--', marker='o')

plt.title('Monthly Revenue vs Expenses')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.legend()

plt.show()

```
