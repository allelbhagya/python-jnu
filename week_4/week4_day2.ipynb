````python
import matplotlib.pyplot as plt


data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [65, 92, 49]
}

df = pd.DataFrame(data)

df

plt.bar(df['Name'], df['Score'])
plt.title('python')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Sales': [143, 120, 255, 310, 150]
}

df = pd.DataFrame(data)

# Line Plot
plt.plot(df['Year'], df['Sales'], marker='x')
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid()
plt.show()
````

````python
# Data
labels = ['mid-sems', 'end sem', 'research paper', 'programming assignment']
sizes = [25, 35, 20, 20]

# Pie Chart
plt.pie(sizes, labels=labels)
plt.title('Mark Distribution')
plt.show()
````

````python
import numpy as np

data = [13,24,42,23,19,98,34,33,18,45,14,58,28,74, 89,89,89,89,89]

plt.hist(data, bins=5)
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid()
plt.show()
````

````python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Scatter Plot
plt.scatter(x, y, color='red')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
````

````python
# Data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Box Plot
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Data Set')
plt.ylabel('Values')
plt.xticks([1, 2, 3], ['Std 1', 'Std 2', 'Std 3'])
plt.grid()
plt.show()
````

````python
data = {
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Production': [150, 200, 250, 300, 400]
}

df = pd.DataFrame(data)

plt.fill_between(df['Year'], df['Production'], color="skyblue", alpha=0.4)
plt.plot(df['Year'], df['Production'], color="Slateblue", alpha=0.6)
plt.title('Production Over Years')
plt.xlabel('Year')
plt.ylabel('Production')
plt.grid()
plt.show()
````
